```python
# Create ../src/cosinor.py from this notebook
%%writefile ../src/cosinor.py
import numpy as np

def cosinor_fit(t_hours, y, period=24.0):
    t = np.asarray(t_hours)
    y = np.asarray(y)
    omega = 2*np.pi/period
    X = np.column_stack([np.ones_like(t),
                         np.cos(omega*t),
                         np.sin(omega*t)])
    beta, *_ = np.linalg.lstsq(X, y, rcond=None)
    M, A, B = beta
    amp = float(np.sqrt(A*A + B*B))
    phi = float(np.arctan2(B, A))
    acrophase_hours = (phi / omega) % period
    yhat = X @ beta
    resid = y - yhat
    return dict(mesor=float(M), amplitude=amp,
                acrophase_hours=float(acrophase_hours),
                yhat=yhat, resid=resid)
```

```python
%%writefile ../src/cosinor.py
import numpy as np

def cosinor_fit(t_hours, y, period=24.0):
    t = np.asarray(t_hours)
    y = np.asarray(y)
    omega = 2*np.pi/period
    X = np.column_stack([np.ones_like(t),
                         np.cos(omega*t),
                         np.sin(omega*t)])
    beta, *_ = np.linalg.lstsq(X, y, rcond=None)
    M, A, B = beta
    amp = float(np.sqrt(A*A + B*B))
    phi = float(np.arctan2(B, A))
    acrophase_hours = (phi / omega) % period
    yhat = X @ beta
    resid = y - yhat
    return dict(mesor=float(M), amplitude=amp,
                acrophase_hours=float(acrophase_hours),
                yhat=yhat, resid=resid)
```

```python
import numpy as np
from src.cosinor import cosinor_fit

t = df["t_hours"].values
y = df["temp_c"].values

result = cosinor_fit(t, y, period=24.0)
result
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(t, y, '.', alpha=0.4, label="Data")
plt.plot(t, result["yhat"], '-', linewidth=2, label="Cosinor fit")
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (°C)")
plt.title("Cosinor Fit to Circadian Rhythm")
plt.legend()
plt.show()
```

```python
resid = result["resid"]
plt.figure(figsize=(12,3))
plt.plot(t, resid, '.', alpha=0.6)
plt.axhline(0, linewidth=1)
plt.xlabel("Time (hours)")
plt.ylabel("Residual (°C)")
plt.title("Residuals")
plt.show()
```

```python
from scipy.signal import periodogram
fs = 12.0  # 5 min sampling = 12 samples/hour
freqs, power = periodogram(y, fs)
plt.figure(figsize=(10,4))
plt.semilogy(freqs, power)
plt.axvline(1/24, linestyle="--")
plt.xlabel("Frequency (1/hour)")
plt.ylabel("Power")
plt.title("Periodogram (peak near 1/24 ≈ 0.0417)")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(t, y, '.', alpha=0.4)
plt.plot(t, result["yhat"], '-', linewidth=2)
plt.xlabel("Hours"); plt.ylabel("°C"); plt.title("Cosinor Fit")
plt.savefig("../report/fig_cosinor_fit.png", dpi=200, bbox_inches="tight")
plt.close()

with open("../report/metrics.txt","w") as f:
    f.write(f"Mesor={result['mesor']:.3f}\nAmplitude={result['amplitude']:.3f}\nAcrophase_hours={result['acrophase_hours']:.2f}\n")
```
